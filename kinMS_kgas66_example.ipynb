```python
from kinms import KinMS
```

```python
import numpy as np
```

```python
scalerad = 3 # arcsec
radius = np.arange(0,1000, 0.1)
sbprof = np.exp(-radius/scalerad)
```

```python
vel = (
    (210) * (2 / np.pi) * np.arctan(radius)
)  # Scaling the maximum velocity to 210 km/s
```

```python
posang = 205 # degrees, east of north to receding side
inc = 50 # degrees
```

```python
# cube dimensions
xsize = 30 # arcsec
ysize = 30 # arcsec
vsize = 500 # km/s
cellsize = 1
dv = 5 # km/s / channel
beamsize = [0.1,0.1,0]
```

```python

kin = KinMS(xsize, ysize, vsize, cellsize, dv, beamSize=beamsize, verbose=True)
```

```python
model=kin.model_cube(inc = inc, sbProf = sbprof,
             sbRad = radius, velProf = vel, posAng = posang, toplot=True)
```
